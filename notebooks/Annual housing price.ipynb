```python
import pandas as pd
import re
import numpy as np
```

```python
import os
```

```python
df1 = pd.read_excel('rent2019-2022.7.xlsx')
```

```python
df1
```

```python
result = []
for i in range(len(df1)):
    x = re.findall('\d+', str(df1.price[i]))
    if len(x) > 1:
        result.append(int(x[0]) + int(x[1])*0.01)
    if len(x) == 1:
        result.append(int(x[0]))
    if len(x) == 0:
        result.append(np.nan)


```

```python
df1.price = result
```

```python
df1
```

```python
re1 = df1.groupby(['postcode','month', 'year']).agg(np.nanmean)
```

```python
mask = [x[2] == 2019 for x in re1.index]
```

```python
re1.loc[mask]
```

```python
yearlist = [2019,2020,2021,2022]

for i in yearlist:
    df_ = df1.loc[df1.year == i]
    re1 = df_.groupby(['postcode','month']).agg(np.nanmean)
    k = re1['price'].unstack(level = -1)
    k = k.dropna(axis = 0, how= 'all')
    print(k)
    k.to_csv(f'{i}_data.csv')
```
