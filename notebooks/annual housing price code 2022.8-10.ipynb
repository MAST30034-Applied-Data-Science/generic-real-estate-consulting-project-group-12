```python
df.iloc[0,7].split(',')[1]
```

```python
df['month'] = df.price
```

```python
monlist = []
for i in range(len(df)):
    re0 = df.iloc[i,7]
    if  type(re0) != str:
        df.iloc[i, 12] = np.nan
        continue
        
    x = re.findall('NowBond', re0)
    if len(x) > 0:
        df.iloc[i, 12] = 'August'
    else:
        d0 = df.iloc[i,7].split(',')
        if len(d0) > 1:
            decom = re.findall('.*ber', d0[1])[0].split(' ')
            df.iloc[i, 12] = decom[2]
```

```python
redf = df.groupby(['postcode', 'month']).agg(np.nanmean)
```

```python
result = redf['price'].unstack(level = -1)
```
